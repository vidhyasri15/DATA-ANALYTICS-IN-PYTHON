```python
### Reorganizing Arrays
```

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
after = before.reshape((4,2))
print(after)
```

```python
import numpy as np

before = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(before)

# Reshape into (2, 3)
after = before.reshape((2, 2, 2))  # Corrected to match the total number of elements
print(after)
```

```python
# Vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2])
#np.hstack([v1,v2,v1,v2])
```

```python
# Horizontal  stack
h1 = np.ones((2,4))
h2 = np.zeros((2,2))

np.hstack((h1,h2))
```

```python
#Scientific Computing
import numpy as np

arr = np.array([1, 4, 9, 16])

print("Square Root:", np.sqrt(arr))
print("Square:", np.square(arr))
print("Exponential:", np.exp([1,2]))
print("Log:", np.log([1,2]))
```

```python
#Array Iteration

arr = np.array([[1,2],[3,4]])

for x in np.nditer(arr):
    print(x)
```

```python
#Joining Arrays
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.concatenate((a,b)))
```

```python
#Splitting Arrays
import numpy as np

arr = np.array([10,20,30,40,50,60])

print(np.array_split(arr,3))
```

```python
#Searching Arrays
import numpy as np

arr = np.array([10,20,30,20,50])

print(np.where(arr==20))
```

```python
#sorting
import numpy as np

arr = np.array([40,10,50,20])

print(np.sort(arr))
```

```python
#FIltering
import numpy as np

arr = np.array([10,15,20,25,30])

print(arr[arr>20])
```
