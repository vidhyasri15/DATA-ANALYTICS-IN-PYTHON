Boolean Indexing

Explanation:
Filter rows using conditions

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["A","B","C"],
    "Marks":[70,85,90]
})

print(df[df["Marks"]>80])
```

Universal Functions

Explanation:
NumPy functions work directly on Pandas objects.

```python
import pandas as pd
import numpy as np

s = pd.Series([1,4,9])

print(np.sqrt(s))
```

Grouping

Explanation:
Group rows and perform aggregate calculations.

```python
import pandas as pd

df = pd.DataFrame({
    "Dept":["IT","IT","HR"],
    "Salary":[50000,60000,45000]
})

print(df.groupby("Dept")["Salary"].mean())
```

Working with Text Data

Explanation:
Use .str methods to manipulate strings.

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["alice","bob"]
})

print(df["Name"].str.upper())
```

Merging

Explanation:
Merge two DataFrames using a common key.

```python
import pandas as pd

df1 = pd.DataFrame({
    "ID":[1,2],
    "Name":["A","B"]
})

df2 = pd.DataFrame({
    "ID":[1,2],
    "Marks":[90,85]
})

print(pd.merge(df1,df2,on="ID"))
```

Joining

Explanation:
Join DataFrames using their index.

```python
import pandas as pd

df1 = pd.DataFrame({"Name":["A","B"]},index=[1,2])
df2 = pd.DataFrame({"Age":[20,25]},index=[1,2])

print(df1.join(df2))
```

Concatenation

Explanation:
Append DataFrames vertically or horizontally.

```python
import pandas as pd

df1 = pd.DataFrame({"A":[1,2]})
df2 = pd.DataFrame({"A":[3,4]})

print(pd.concat([df1,df2]))
```

Date and Time Operations

Explanation:
Convert strings into datetime objects and extract components.

```python
import pandas as pd

df = pd.DataFrame({
    "Date":["2026-06-30","2026-07-01"]
})
df["Date"] = pd.to_datetime(df["Date"])
print(df["Date"].dt.day)
print(df["Date"].dt.month)
print(df["Date"].dt.year)
```
