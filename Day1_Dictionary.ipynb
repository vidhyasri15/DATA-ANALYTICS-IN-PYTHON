# **Mini Challenge**

 write a program that:


1.Creates a dictionary of 5 students and their marks.

2.Calculates the average marks.

3.Finds the highest and lowest scorer.

4.Displays all students scoring above the average.

5.Allows the user to search for a student by name.

6.Updates a student's marks and displays the revised average.

```python
# Activity: Dictionary – Student Grades Management

grades = {
    "Alice": 85,
    "Bob": 78,
    "Charlie": 92
}

while True:
    print("\n========== STUDENT GRADE MANAGEMENT ==========")
    print("1. Display all students and grades")
    print("2. Calculate average grade")
    print("3. Find highest scorer")
    print("4. Find lowest scorer")
    print("5. Add a new student")
    print("6. Update a student's grade")
    print("7. Remove a student")
    print("8. Search for a student")
    print("9. Display students scoring above 80")
    print("10. Display students scoring above average")
    print("11. Assign Letter Grades (A, B, C)")
    print("12. Count total number of students")
    print("13. Exit")

    choice = int(input("\nEnter your choice (1-13): "))

    if choice == 1:
        print("\nStudent Grades")
        for student, grade in grades.items():
            print(student, ":", grade)

    elif choice == 2:
        average = sum(grades.values()) / len(grades)
        print("\nAverage Grade =", average)

    elif choice == 3:
        highest = max(grades, key=grades.get)
        print("\nHighest Scorer:", highest)
        print("Marks:", grades[highest])

    elif choice == 4:
        lowest = min(grades, key=grades.get)
        print("\nLowest Scorer:", lowest)
        print("Marks:", grades[lowest])

    elif choice == 5:
        name = input("Enter student name: ")
        mark = int(input("Enter marks: "))
        grades[name] = mark
        print("Student added successfully.")

    elif choice == 6:
        name = input("Enter student name to update: ")
        if name in grades:
            mark = int(input("Enter new marks: "))
            grades[name] = mark
            print("Marks updated successfully.")
        else:
            print("Student not found.")

    elif choice == 7:
        name = input("Enter student name to remove: ")
        if name in grades:
            del grades[name]
            print("Student removed successfully.")
        else:
            print("Student not found.")

    elif choice == 8:
        name = input("Enter student name to search: ")
        if name in grades:
            print(name, "scored", grades[name], "marks.")
        else:
            print("Student not found.")

    elif choice == 9:
        print("\nStudents scoring above 80")
        found = False
        for student, mark in grades.items():
            if mark > 80:
                print(student, ":", mark)
                found = True
        if not found:
            print("No student scored above 80.")

    elif choice == 10:
        average = sum(grades.values()) / len(grades)
        print("\nAverage Grade =", average)
        print("Students scoring above average:")
        found = False
        for student, mark in grades.items():
            if mark > average:
                print(student, ":", mark)
                found = True
        if not found:
            print("No student scored above average.")

    elif choice == 11:
        print("\nLetter Grades")
        for student, mark in grades.items():
            if mark >= 90:
                grade = "A"
            elif mark >= 75:
                grade = "B"
            else:
                grade = "C"

            print(student, ":", mark, "-", grade)

    elif choice == 12:
        print("\nTotal Number of Students =", len(grades))

    elif choice == 13:
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice! Please enter a number between 1 and 13.")
```

```python
#Exercise – Contact Book with Dictionaries
# Contact Book Program
contacts = {}

def add_contact(name, phone):
    contacts[name] = phone
    print(f"Added {name} : {phone}")

def view_contacts():
    for name, phone in contacts.items():
        print(name, ":", phone)

def search_contact(name):
    if name in contacts:
        print(name, ":", contacts[name])
    else:
        print("Contact not found.")

# Example usage
add_contact("Ankit", "12345")
add_contact("Ani", "98765")
view_contacts()
search_contact("Alice")
```

```python
import random
import math
import string
from datetime import datetime
import os
import sys

# Random Module
print("----- RANDOM MODULE -----")
print("Random Number:", random.randint(1, 10))
print("Random Fruit:", random.choice(["Apple", "Banana", "Cherry"]))

# Math Module
print("\n----- MATH MODULE -----")
print("Square Root of 25:", math.sqrt(25))
print("Sine of 90°:", math.sin(math.pi / 2))

# String Module
print("\n----- STRING MODULE -----")
print("ASCII Letters:", string.ascii_letters)
print("Digits:", string.digits)

# Datetime Module
print("\n----- DATETIME MODULE -----")
current_date = datetime.now()
print("Current Date & Time:", current_date.strftime("%Y-%m-%d %H:%M:%S"))

# OS Module
print("\n----- OS MODULE -----")
print("Current Directory:", os.getcwd())

print("\nFiles and Folders:")
for file in os.listdir(os.getcwd()):
    print(file)

# Sys Module
print("\n----- SYS MODULE -----")
print("Command Line Arguments:", sys.argv)
print("\nPython Version:")
print(sys.version)
```
